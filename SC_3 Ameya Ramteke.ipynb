6: Plotting Histogram of color channel

```python
import matplotlib.pyplot as plt
image4 = cv2.imread(r'AdobeStock_194578054.jpeg')
for i,col in enumerate(['r','g','b']):
    hist = cv2.calcHist([image4],[i],None,[256],[0,256])
    plt.title("colour Histogram")
    plt.xlabel("Bin")
    plt.ylabel("Pixels")
    plt.plot(hist, color=col)
    plt.xlim([0,256])

plt.show()
plt.imshow(image4)

```

Image to Grayscale

```python
from matplotlib import pyplot as plt
import cv2
img = cv2.imread(r'AdobeStock_194578054.jpeg')
R, G, B = img[:,:,0], img[:,:,1], img[:,:,2]
imgGray = 0.2989 * R + 0.5870 * G + 0.1140 * B #Conversion of an image to grayscale
plt.hist(imgGray.ravel(), 256, (0, 256))
plt.show()
plt.imshow(imgGray, cmap='gray')
```

Building Pyramids

```python
img = cv2.imread(r'AdobeStock_194578054.jpeg')
layer = img.copy()
for i in range(4):
 plt.subplot(2, 2, i + 1)
 layer = cv2.pyrDown(layer)
 plt.imshow(layer)
 cv2.imshow("str(i)", layer)
 cv2.waitKey(0)
cv2.destroyAllWindows()
```
