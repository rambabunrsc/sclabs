```python
import geopandas as gpd
from pyproj import crs
import matplotlib.pyplot as plt
```

```python
countries = gpd.read_file("countries.shp")
airports = gpd.read_file("airports.shp")
railroads =gpd.read_file("railroads.shp")
```

```python
#creating static maps
fig, ax = plt.subplots(figsize = (15,10))
#visualising the layer into 5 classes using Quantiles classification
countries.plot(ax=ax, column ="LABELRANK",linewidth=0.03,cmap="Spectral",scheme="quantiles", k=5, alpha =0.9)
#adding the airports point shape layer
airports.plot(ax=ax, color ="blue")
#adding the rail roads layer
railroads.plot(ax=ax, color ="yellow", linewidth=1.5)
plt.tight_layout()
#save the figure as png with specified resolution
out = "staticmap.png"
plt.savefig(out, dpi=350)
```

```python
fig, ax= plt.subplots(figsize =(6,4))
#visualise the country into 5 classes of quantiles classification
#using the attribute column population rank of countries wise
countries.plot( ax= ax, column = "POP_RANK", linewidth = 0.03, 
               cmap ="Spectral", alpha = 0.9, legend = True,legend_kwds={'label':"Rank"})
plt.tight_layout()
```
