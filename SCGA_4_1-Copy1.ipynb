```python
import rasterio
from os import listdir
from os.path import isfile, join
```

```python
path = 'D:/rambabu_r@nrsc.gov.in/STUDY/MTECH/IIST GeoInformatics Text books/1 SEM Books/SCGA&LAB/LABS/SCGA_LAB_4/R20m'
```

```python
files = [f for f in listdir(path) if isfile(join(path,f))]
```

```python
print(files[1:14])
```

```python
# reading the metadata of first file
with rasterio.open(path+'/'+files[1]) as src0:
    meta = src0.meta
    meta.update(count = len(files[1:14]))

      
```

```python
#read each layer and write it to stack
with rasterio.open('stack.tiff', "w",**meta) as dst:
        for id, layer in enumerate (files[1:14], start = 1):
            print(path+'/'+layer)
            with rasterio.open(path+'/'+layer) as src1:
                dst.write_band(id, src1.read(1))
                print('stacked.tiff')
```

```python
stack = rasterio.open('stack.tiff')
```

```python
stack.read()
```

```python
from rasterio.plot import show_hist
show_hist(stack, bins=50, lw=0.0, stacked=False, alpha=0.7,
        histtype='stepfilled', title="Histogram")
```
