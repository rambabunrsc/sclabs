```python
#histogram
from matplotlib import pyplot as plt
import cv2
img = cv2.imread("moon.jpg")
R, G, B = img[:,:,0], img[:,:,1], img[:,:,2]
imgGray = 0.2989 * R + 0.5870 * G + 0.1140 * B #Conversion of an image to grayscale
plt.hist(imgGray.ravel(), 256, (0, 256))
plt.show()
plt.imshow(imgGray, cmap='gray')
```

```python
#pyramids
img = cv2.imread("moon.jpg")
layer = img.copy()
for i in range(4):
 plt.subplot(2, 2, i + 1)
 layer = cv2.pyrDown(layer)
 plt.imshow(layer)
 cv2.imshow("str(i)", layer)
 cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#image properties
from PIL.ExifTags import TAGS
image = Image.open('moon.jpg')
img = cv2.imread('moon.jpg')

#dimensions
dimensions = img.shape
height = img.shape[0]
width = img.shape[1]
channel = img.shape[2]
print(image.format)
print(image.filename)
print(image.format_description)
print("date type",type(img))
print('Image dimension:',dimensions)
print('Image:Height : ',height)
print('Image Width : ',width)
print('Print the number of channel: ',channel)
```

```python
#metadata
exifdata = image.getexif()
for tagid in exifdata:
    tagname = TAGS.get(tagid,tagid)
    value = exifdata.get(tagid)
    print(f"{tagname:25}:{value}")
```

```python
#splitting
import cv2
import matplotlib.pyplot as plt
image2 = cv2.imread('moon.jpg')
b,g,r = cv2.split(image2)
cv2.imshow('image Blue',b)
cv2.waitKey(0)
cv2.imshow('image Green',g)
cv2.waitKey(0)
cv2.imshow('image red',r)
cv2.waitKey(0)
cv2.destroyAllWindows()

plt.subplot(1,1,1)
cv2.imshow('image blue',b)
cv2.waitKey(0)
plt.subplot(1,1,1)
cv2.imshow('image green',g)
cv2.waitKey(0)
plt.subplot(1,1,1)
cv2.imshow('image red',r)
cv2.waitKey(0)
```

```python
#Merging
image2 = cv2.imread('moon.jpg')
b,g,r = cv2.split(image2)
cv2.imshow('image Blue',b)
cv2.waitKey(0)
cv2.imshow('image Green',g)
cv2.waitKey(0)
cv2.imshow('image red',r)
cv2.waitKey(0)
cv2.destroyAllWindows()
image3 = cv2.merge([b,g,r])
cv2.imshow('merged',r)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#resize
img = cv2.imread('moon.jpg')
layer = img.copy()
for i in range(4):
 plt.subplot(2, 2, i + 1)
 layer = cv2.pyrUp(layer)
 plt.imshow(layer)
 cv2.imshow("str(i)", layer)
 cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#ROI
from skimage import io
import numpy as np
img_o = cv2.imread('moon.jpg')
img_skimg = io.imread('moon.jpg')
#print rectangle points of selected roi
roi_cropped = img_o[100:300,200:600] #xmin:ymin,xmax:ymax
#show cropped image
cv2.imshow("ROI", roi_cropped)
cv2.imwrite("crop.jpeg",roi_cropped)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
